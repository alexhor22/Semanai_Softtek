```python
def parseDimentions(string):
    values = string.split(" ")
    angle = float(values[2])
    mayor_axis = float(values[0])
    minor_axis = float(values[1])
    width = math.sqrt(math.pow(mayor_axis*math.sin(angle+(math.pi)/2),2)+math.pow(minor_axis*math.sin(angle),2))*2
    height = math.sqrt(math.pow(mayor_axis*math.cos(angle+(math.pi)/2),2)+math.pow(minor_axis*math.cos(angle),2))*2
    x = float(values[3])-width/2
    y = float(values[4])-height/2
    return int(x),int(y),int(width),int(height)
```

```python
def readEipse(file):
    with open(file,"r") as f:
        array = f.read().splitlines()
        longitud = len(array)
        i = 0
        dict_array = []
        while (i < longitud):
            dict = {"name":"","annotations":[]}
            if (re.search("big_img",array[i])):
                try:
                    dict["name"] = "{}.jpg".format(array[i])
                    img = mpimg.imread("dataset/"+array[i]+".jpg")
                    (h,w,_) = img.shape #Obtener dimensiones de la imagen

                    n = int(array[i+1])
                    i=i+2
                    #fig,ax = plt.subplots(1)
                    #ax.imshow(img)
                    for n in range(n):
                        x,y,width,height = parseDimentions(array[i])
                        (x,width) = (0,width+x) if x < 0 else (x,width)
                        (y,heigth) = (0,height+y) if y < 0 else (y,height)
                        width = w-x-1 if x+width >= w else width
                        height = h-y-1 if y+height >= h else heigth
                        if (x + width < w and y + height < h):
                            #rect = patches.Rectangle((x,y),width,height,linewidth=1,edgecolor='r',facecolor='none')
                            #ax.add_patch(rect)
                            dict["annotations"].append([x,y,width,height])
                            dict["size"] = {"height":h,"width":w}
                            #plt.imshow(img[y-height:y+height,x-width:x+width])
                            #plt.show()
                        else:
                            #print("{},{},{}".format(w,x,width))
                            #print("{},{},{}".format(h,y,height))
                            print("out of bounds")
                        i=i+1
                except:
                    print("{} not found".format(array[i]))
                    i=i+1
                #plt.show()
                dict_array.append(dict.copy())
            else:
                i=i+1
        return dict_array
```
